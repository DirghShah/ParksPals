```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 100)
```

## Load data

Read the Excel files. These are unmodified files from [Outscraper](https://outscraper.com/)'s scraped output. They contain Google reviews for leasing companies, apartment complexes, and student housing for three different campustowns.

- Brigham Young University (Provo, UT)
- Penn State University (University Park, PA)
- University of Illinois at Urbana Champaign (Champaign, IL)

```python
df_byu_b = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/byu-businesses.csv')
df_byu_r = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/byu-reviews.csv')
df_psu_b = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/psu-businesses.csv')
df_psu_r = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/psu-reviews.csv')
df_uiuc_b = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/uiuc-businesses.csv')
df_uiuc_r = pd.read_csv('https://github.com/bdi475/datasets/raw/main/campustowns-leasing-company-reviews/uiuc-reviews.csv')
```

```python
display(df_byu_b.head(2))
display(df_byu_r.head(2))
display(df_psu_b.head(2))
display(df_psu_r.head(2))
display(df_uiuc_b.head(2))
display(df_uiuc_r.head(2))
```

```python
df_byu_b['campus'] = 'Brigham Young University'
df_psu_b['campus'] = 'Penn State University Park'
df_uiuc_b['campus'] = 'University of Illinois Urbana-Champaign'
```

```python
# businesses
df_b = pd.concat([df_byu_b, df_psu_b, df_uiuc_b])

# reviews
df_r = pd.concat([df_byu_r, df_psu_r, df_uiuc_r])
```

```python
print(f'df_b.shape={df_b.shape}')
print(f'df_r.shape={df_r.shape}')
```

```python
columns_to_select1 = [
    'campus',
    'place_id',
    'name',
    'site',
    'category',
    'borough',
    'street',
    'city',
    'postal_code',
    'state',
    'latitude',
    'longitude',
    'verified',
]

df_b = df_b[columns_to_select1].copy()

df_b.head()
```

```python
df_r.head(3)
```

```python
columns_to_select2 = [
    'place_id',
    'review_id',
    'author_id',
    'author_title',
    'review_text',
    'review_rating',
    'review_img_url',
    'review_datetime_utc',
    'owner_answer',
    'owner_answer_timestamp_datetime_utc',
    'review_likes'
]

df_r = df_r[columns_to_select2].copy()

df_r.head()
```

```python
print(f'df_b.shape={df_b.shape}')
print(f'df_r.shape={df_r.shape}')
```

```python
df_r[['review_datetime_utc', 'owner_answer_timestamp_datetime_utc']].dtypes
```

```python

df_r['review_datetime_utc'] = pd.to_datetime(df_r['review_datetime_utc'], utc=True)
df_r['owner_answer_timestamp_datetime_utc'] = pd.to_datetime(df_r['owner_answer_timestamp_datetime_utc'], utc=True)
```

```python
df_b.to_csv('businesses.csv', index=None)
df_r.to_csv('reviews.csv', index=None)
```

Merging Database on place_id for future analysis

```python
df_br = pd.merge(
    left=df_b,
    right=df_r,
    on='place_id',
    how='left')

df_br.head(2)
```

```python
df_campus_avgratings = df_br.groupby(['campus','name']).agg({'review_rating': 'mean'}).reset_index()

df_campus_avgratings
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
average_rating_per_campus = df_br.groupby('campus')['review_rating'].mean().reset_index()
fig = px.bar(average_rating_per_campus, x='campus', y='review_rating',
                        title='Average Review Rating per Campus',
                        labels={'review_rating': 'Average Rating', 'campus': 'Campus'}, color = 'campus')

fig.show()
```

```python
business_category_distribution_grouped = df_br.groupby(['campus', 'category']).size().reset_index(name='count')

fig_category_distribution_grouped = px.bar(business_category_distribution_grouped,
                                           x='campus', y='count', color='category',
                                           barmode='group',
                                           title='Business Category Distribution per Campus (Grouped)',
                                           labels={'count': 'Number of Businesses', 'category': 'Business Category'})

# Displaying the figure
fig_category_distribution_grouped.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

df_r['review_text'] = df_r['review_text'].astype(str)
combined_text = " ".join(df_r['review_text'])

stopwords = set(STOPWORDS)

wordcloud = WordCloud(width = 1000, height = 1000,
                      background_color ='white',
                      stopwords = stopwords,
                      min_font_size = 10).generate(combined_text)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
df_uiuc_r['review_datetime_utc'] = pd.to_datetime(df_uiuc_r['review_datetime_utc'])
df_uiuc_r['year'] = df_uiuc_r['review_datetime_utc'].dt.year
df_uiuc_r['month_name'] = df_uiuc_r['review_datetime_utc'].dt.strftime('%b')
df_uiuc_r['month'] = df_uiuc_r['review_datetime_utc'].dt.month
# Current year
current_year = 2023

# Filtering the data to include only the last three years
df_recent_years = df_uiuc_r[df_uiuc_r['year'] >= current_year - 2]

# Grouping by year and month name
df_recent_monthly_avg = df_recent_years.groupby(['year', 'month_name','month'])['review_rating'].mean().reset_index()

# Sorting the data by year and month for consistent plotting
df_recent_monthly_avg.sort_values(by=['year', 'month'], inplace=True)

# Creating a time series plot with the x-axis as month names and different lines for each of the last three years
fig_recent_years_months = px.line(df_recent_monthly_avg,
                                  x='month_name',
                                  y='review_rating',
                                  color='year',
                                  title='UIUC Review Rating Change Over Time (Monthly, Last 3 Years)',
                                  labels={'review_rating': 'Average Rating', 'month_name': 'Month', 'year': 'Year'})

# Displaying the plot
fig_recent_years_months.show()
```
